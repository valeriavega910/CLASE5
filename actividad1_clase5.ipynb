4 -Ingresar dos números por teclado a y b. Si a es igual a b, imprimir 1
si a es mayor que b, imprimir 2 sino imprimir 3.

```python
a=int(input("Ingrese un valor para a"))
b=int(input("Ingrese un valor para b"))
if a==b:
    print("1")
elif a>b:
    print("2")
else:
    print("3")
```

9 -Ingresar por teclado tres
números y mostrar por pantalla el mayor de los tres.

```python
numero1=int(input("Ingrese numero 1"))
numero2=int(input("Ingrese numero 2"))
numero3=int(input("Ingrese numero 3"))
if numero1>numero2 and numero1>numero3:
    print("El primer numero",numero1,"es el mayor")
elif numero2>numero1 and numero2>numero3:
    print("El segundo numero",numero2,"es el mayor")
else:
    print("El tercero",numero3,"es el mayor de los tres")
```
