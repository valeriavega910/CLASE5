- Ciclo For in
- Ciclo While
- Break
- Continue
Los bucles son estructuras fundamentales en programación que permiten ejecutar un bloque de código de manera repetitiva hasta que se cumpla una condición específica. Estas estructuras son escenciales para automatizar tareas repetitivas y controlar el flujo de un programa.

```python
# Iterear lista con For

colores=["rojo","amarillo","verde","marron","violeta"]
for p in colores:
    print(f"el color ahora es igual a {p}")
```

```python
# función range()
for numeros in range(0,120,5):
    print(numeros)
```

```python
# Forma de recorrer una lista a traves de su  indice

for color in enumerate(colores):
    print(f"El color en la posición {color[0]} es {color[1]}")
```

```python
# Forma de recorrer una lista a traves de su  indice

for color in enumerate(colores):
    print(color)
```

```python
# Recorrer el diccionario a traves del indice

diccionario_val = {"NOMBRE":"Valeria",
                   "EDAD":40,
                   "CARRERA":"Tec. en Ciencia de Datos e I.A.",
                   "SEXO":"femenino",
                   "PROFESION":"docente",
                   "VEHICULO":{"marca":"Nissan",
                               "modelo":"kicks",
                               "año":2024}}
print(diccionario_val)
```

```python
for v in diccionario_val.items():
    print(v)
```

```python
for v in diccionario_val.keys():
    print(v)
```

```python
for v in diccionario_val.values():
    print(v)
```

```python
# como mostrar las claves
for keys in diccionario_val:
    print(keys)
```

```python
# Acceder a una clave específica
print(diccionario_val["VEHICULO"])
```

```python
# Cambiar el flujo normal de un ciclo con continue, encuentra un elemento. lo salta y 
# y continua hasta el final
frutas=["banana","manzana","kiwi","ciruela","uva","sandia"]
for fruta in frutas:
    if fruta=="ciruela" or fruta=="pera":
        continue
    print(f"Voy a comer una {fruta}")
```

```python
# Con el break encuentra el elemento y se detiene.
for fruta in frutas:
    if fruta=="uva":
        break
    print(f"Voy a comer una {fruta}")
```
