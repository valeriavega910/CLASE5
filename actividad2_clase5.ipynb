```python
# 5. Imprimir los números del 1 al 100, para los múltiplos de 3, imprimir “múltiplo de 3”, para los múltiplos de 5 imprimir “múltiplo de 5” y para los múltiplos de 3 y de 5 imprimir “múltiplo de 3 y 5”.
multiplo3=[]
multiplo5=[]
multiplo3y5=[]

for t in range(1,101):
    if t % 3 == 0 and t % 5 == 0:
        multiplo3y5.append(t)
    elif t % 5 == 0:
        multiplo5.append(t)
    elif t % 3 == 0:
        multiplo3.append(t)
print("Los multiplos de 3 son ", multiplo3)
print("Los multiplos de 5 son ", multiplo5)
print("Los multiplos de 3 y 5 son ", multiplo3y5)
```

```python
# 6. Ingresar un número por teclado e imprimir si es número primo o no.
def es_primo(n): #funcion
    if n < 2:
        return False
    for i in range(2,n):
        if n % i == 0:
            return False
    return True

numero=int(input("ingrese un numero: "))
if es_primo(numero):
    print("el numero es primo")
else:
    print("el numero no es primo")
```

```python
# 7. partir de una lista vacía, utilizar un ciclo while para cargar allí números negativos del -20 al -1
numeros_negativos = []
i = -20
while i < 0:
    numeros_negativos.append(i)
    i += 1
    print(numeros_negativos)
```

```python
# 11. Imprimir los números primos del 1 al 100
def es_primo(n): #funcion
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

for i in range(1,101):
    if es_primo(i):
        print(i)
```

```python
# 15.Dada una lista de marca de vehículos
# marcas =[‘Ford’,’Renault’,’Peugeot’,’Audi’,’Toyota’], recorrer la lista con un bucle for 
# y en caso de que la marca sea ‘Audi’ salir del bucle.
marcas = ['Ford', 'Renault', 'Peugeot', 'Audi', 'Toyota']
for marca in marcas:
    if marca == 'Audi':
        break
    print(marca)
```
